# Reproduce a plot; create a bar plot

# Specific goal: Create a chart with 2 different Y-axes each with a different scale and each with a different chart type (bar plot & line)

```python
#Solution: twinx() function to create a second axis with a different scale

import matplotlib.pyplot as plt
fig,ax=plt.subplots() # create canvas and axes
year= ["2015", "2016", "2017", "2018", "2019"] # Independent Variable  = year
employees= [24000, 30000, 22000, 22000, 22000] # Dependent Variable 1 (DV1) = employee number
total_employee_costs= [1800000, 1600000,1700000, 1700000, 2100000] # DV 2 = employee costs
ax.bar (year, employees, color='yellow') # create bar chart for DV1
ax.set_title('Number of Employees & Total Employee Costs for 2015-2019')
ax.set_ylabel('Number of Employees')

ax2=ax.twinx() # create second axis

ax2.plot(year,total_employee_costs)
ax2.set_ylabel('Total Employee Costs')
```
